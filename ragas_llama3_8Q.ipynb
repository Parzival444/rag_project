```python
import os
from ragas import evaluate
from ragas.metrics import (
    answer_correctness,
    context_precision,
    context_recall,
    faithfulness,
    answer_relevancy,
)
from datasets import Dataset
metrics = [
    answer_relevancy,
    faithfulness,
    context_precision,
    context_recall,
    answer_correctness,
]
from collections import defaultdict

# Store per-question scores
all_scores = defaultdict(list)

print("Per-sample RAGAS evaluation:\n")

for i, sample in enumerate(sample_data):
    # Convert single sample to Dataset format
    single_dataset = Dataset.from_dict({
        "question": [sample["question"]],
        "retrieved_contexts": [sample["retrieved_context"]],
        "response": [sample["generated_answer"]],
        "ground_truth": [sample["ground_truth_answer"]],
    })

    result = evaluate(single_dataset, metrics)

    print(f"Sample {i+1}: {sample['question']}")
    for metric in metrics:
        name = metric.name
        value = result[name]
        score = value[0] if isinstance(value, list) else value  # handle ragas 0.0.3 vs newer versions
        all_scores[name].append(score)
        print(f"  {name}: {score:.4f}")
    print()
    
print("Average Scores Across All Samples:\n")
for name, scores in all_scores.items():
    valid_scores = [s for s in scores if s is not None and not (isinstance(s, float) and (s != s))]  # filter NaN
    avg = sum(valid_scores) / len(valid_scores) if valid_scores else 0
    print(f"{name}: {avg:.4f}")
```
